```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
original = np.arange(-100,100)
original
```

```python
padronizado = (original - np.min(original))/np.max(original)-np.min(original)
padronizado
```

```python
normalizado = (original - np.mean(original))/np.std(original)
normalizado
```

```python
plt.figure(figsize=(18,10))
plt.scatter(original, padronizado, marker='*')
plt.scatter(original, normalizado, marker='.')
plt.legend(['Padronizados', 'Normlizados'])
plt.show()
```
